```python
import numpy as np
import pandas as pd 
train = pd.read_csv('../data/train_data.csv')
test = pd.read_csv('../data/test_data.csv')
train
```

```python

train.fillna(method='ffill',inplace=True)
```

```python
# do some preprocessing :)
from hazm import stopwords_list
from hazm import word_tokenize
test.isnull().sum()
def removeStopWords(x):
    list1=[]
    for i in x:
        if i not in stopwords_list():
            list1.append(i)
    return list1
train['proccessBio'] = train['biography'].apply(lambda x : removeStopWords(x))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.feature_extraction.text import CountVectorizer
count_vect1 = TfidfVectorizer(stop_words='english',input='content',decode_error='replace',strip_accents='unicode',min_df=2)
count_vect2 = TfidfVectorizer(stop_words='english',input='content',decode_error='replace',strip_accents='unicode',min_df=2)
count_vect3 = TfidfVectorizer(stop_words='english',input='content',decode_error='replace',strip_accents='unicode',min_df=2)



import emoji 
train['newName'] = train['fullname'].apply(lambda x : emoji.demojize(x))
train['newUser'] = train['username'].apply(lambda x : emoji.demojize(x))
train['newbio'] = train['biography'].apply(lambda x : emoji.demojize(x))
finalFullName = count_vect1.fit_transform(train['newName'].values)
finalUserName = count_vect2.fit_transform(train['newUser'].values)
finalBiography =  count_vect3.fit_transform(train['newbio'].values)
name = pd.DataFrame(finalFullName.toarray())
user = pd.DataFrame(finalUserName.toarray())
bio =  pd.DataFrame(finalBiography.toarray())
train = pd.concat([train,name,user,bio],axis=1)
train['gender']= train['gender'].replace(['man','woman'],[1,0])
train.drop(['fullname' , 'username' , 'biography','newbio' , 'newUser' , 'newName' ],axis=1, inplace=True)
```

```python
#for test data 
test['newName'] = test['fullname'].apply(lambda x : emoji.demojize(x))
test['newUser'] = test['username'].apply(lambda x : emoji.demojize(x))
test['newbio'] = test['biography'].apply(lambda x : emoji.demojize(x))
finalFullName = count_vect1.transform(test['newName'].values)
finalUserName = count_vect2.transform(test['newUser'].values)
finalBiography =  count_vect3.transform(test['newbio'].values)
name = pd.DataFrame(finalFullName.toarray())
user = pd.DataFrame(finalUserName.toarray())
bio =  pd.DataFrame(finalBiography.toarray())
test = pd.concat([test,name,user,bio],axis=1)
test.drop(['fullname' , 'username' , 'biography','newbio' , 'newUser' , 'newName' ],axis=1, inplace=True)
```
